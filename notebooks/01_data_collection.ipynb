```python
# Define Parameters
# Set date range to be from 2000-2025
years = list(range(2000, 2026))

# Define genres of music for dataset
genres = [
    "pop",
    "rock",
    "hip-hop",
    "rap",
    "electronic",
    "dance",
    "indie",
    "jazz",
    "classical",
    "country",
    "latin",
    "rnb"
]

# Define letters to randomly search
letters = list("abcdefghijklmnopqrstuvwxyz")

# Create track storage list
tracks = []

results = sp.search(q="pop year:2015", type="track", limit=5, market="GB")

for track in results["tracks"]["items"]:
    print(track["name"], "-", track["artists"][0]["name"])
```

```python
# Collect song data
# Loop through parameters to find songs
tracks = []

for year in years:
    
    for genre in genres:
        
        for offset in range(0, 200, 50):
            
            query = f"{genre} year:{year}"
            
            try:
                
                results = sp.search(
                    q=query,
                    type="track",
                    limit=50,
                    offset=offset,
                    market="GB"
                )
                
                items = results["tracks"]["items"]
                
                if len(items) == 0:
                    continue
                
                for track in items:
                    print(track["name"])
                    
                    tracks.append({
                        "track_name": track["name"],
                        "artist": track["artists"][0]["name"],
                        "artist_id": track["artists"][0]["id"],
                        "track_id": track["id"],
                        "popularity": track["popularity"],
                        "release_date": track["album"]["release_date"],
                        "duration_ms": track["duration_ms"],
                        "explicit": track["explicit"]
                    })
                
                time.sleep(0.2)
            
            except:
                continue
```

```python
df_tracks = pd.DataFrame(tracks)

print("Raw tracks collected:", len(df_tracks))
```
