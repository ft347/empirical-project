## 1. Introduction

This project aimed to predict the listing prices of Airbnbs in London using machine learning techniques. By analysing various property features, such as location (latitude and longitude) and room types, models were developed to better understand the primary drivers of rental costs in London

## 2. Key Findings

- **Location vs Property Type**: While geographic location (latitude and longitude) are significant in determining listing price, the type of room (whether a listing is a private room or not) was the single most influential factor in determining price
  
- **Property Features**: There is a strong positive correlation between price and the number of bathrooms and bedrooms. This indicates that property capacity and privacy type are the primary drivers of value for guests in London

## 3. Model Performance

The Random Forest model generally outperformed the Linear Regression by capturing non-linear relationships between variables like location and price

```python
# Import library
import pandas as pd

results = pd.DataFrame({
    "Model": ["Linear Regression", "Random Forest"],
    "R2": [0.621476844055324, 0.7968027082032036]
})

results["R2"] = results["R2"].round(3)

results.style.hide(axis='index')
```

## 4. Feature Importance

Based on the Random Forest Model, these features had the highest impact on listing price:

1. **Private Room**: This is by far the most significant features and so indicates a large gap in listing prices between shared/private rooms and entire rental units

2. **Bathrooms**: The number of bathrooms featured in a property was the most significant property specific feature, showing strong correlation between a greater number of bathrooms and higher price points

3. **Location (Latitude/Longitude)**: The proximity of a property to central London remains an important factor in determining listing price

4. **Availability**: Properties available for more days in the year (available 365 days) tend to have different pricing structures compared to those only available occasionally

## 5. Limitations

- **Date of Data**: The dataset used only represents a snapshot in time and so doesn't fully capture seasonal trends, such as how listing prices change in summer vs in winter
  
- **Property Features**: The dataset lacks high-value property features, such as air conditioning and hot tubs, etc., which might explain outliers in price

- **Log Price Transformation**: Models were trained using log_price in order to handle skewness, which can make direct interpretations of errors in price more complex

## 6. Improvements

- **External Data**: Merging the dataset with additional data could help address some of the limitations and also improve the accuracy of the model conclusions. For example, including data over different periods of time to account for seasonal differences or merging with tube station proximity to improve the importance score of location
  
- **Rating Analysis**: Rather than simply using a numeric rating on the property, the description or reviews could be analysed to gauge a rating of the quality of the host

- **Hyperparameter Tuning**: The root mean squared error returned by the Random Forest model could be reduced by fine tuning the hyperparameters, for example by allowing it to search for every possible combination, rather than limiting the model

## 7. Conclusion

The analysis of this dataset successfully identifies that the type of listing, the bathroom count and the location are more predictive of price than the popularity of a property (the number of reviews). This analysis can be used by hosts looking to optimise their listing price, suggesting that they should focus on the privacy of the room and the number of available facilities. While for guests, the data suggests that moving slightly further out of Central London offers better value without necessarily sacrificing the quality of the property.
