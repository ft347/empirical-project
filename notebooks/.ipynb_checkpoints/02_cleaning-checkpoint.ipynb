# BEE2041 Empirical Project: What property and location factors determine Airbnb listing prices in London?
## 02_cleaning

```python
# Import Libraries
import pandas as pd
import numpy as np

# Import Data
df = pd.read_csv("../data/raw/london.csv")

# View first 5 rows
df.head()
```

```python
# View Columns
df.columns

# List Desired Columns
cols = [
    "price",
    "accommodates",
    "bedrooms",
    "beds",
    "bathrooms",
    "property_type",
    "room_type",
    "latitude",
    "longitude",
    "review_scores_rating",
    "number_of_reviews",
    "reviews_per_month",
    "availability_365",
    "host_is_superhost",
    "host_listings_count",
    "instant_bookable"
]

# Create new DataFrame with only the desired columns
df_1 = df[cols].copy()

# View first 5 rows of new DataFrame
df_1.head()
```

```python
# Change price from a string to a float
df_1["price"] = df_1["price"].replace(r"[\$,]", "", regex=True).astype(float)

# View change
df_1["price"].head()
```

```python
# Check for missing values
print(df_1.isnull().sum())

# Drop rows with missing values
df_1 = df_1.dropna()

# Check for missing values again
print(df_1.isnull().sum())
```

```python
# Fix boolean variables
df_1["host_is_superhost"] = df_1["host_is_superhost"].map({"t": 1, "f": 0})
df_1["instant_bookable"] = df_1["instant_bookable"].map({"t": 1, "f": 0})

# View first 10 rows
df_1.head(10)
```

```python
# Check number of properties priced above $1000
print(len(df_1[df_1["price"]>1000]))

# Remove these extreme outliers
df_1 = df_1[df_1["price"] <= 1000]

# Check number of properties priced above $1000 again
print(len(df_1[df_1["price"]>1000]))
```

```python
# Simplify property_type
df_1["property_type_simple"] = df_1["property_type"].apply(
    lambda x: "Entire" if "Entire" in x
    else "Private Room" if "Private room" in x
    else "Shared Room" if "Shared room" in x
    else "Other"
)

# Drop original property_type column
df_1 = df_1.drop("property_type", axis=1)

# View first 5 rows
df_1.head()
```

```python
# Encode category columns
df_1 = pd.get_dummies(df_1, columns=[
    "property_type_simple",
    "room_type"
], drop_first=True, dtype=int)

# Show first 10 rows
df_1.head(10)
```

```python
# Log transform price
df_1["log_price"] = np.log(df_1["price"])

# Check DataFrame
print(df_1.info())
print(df_1.describe())
```

```python
# Save cleaned DataFrame
df_1.to_csv("../data/cleaned/london_clean.csv", index=False)
```
